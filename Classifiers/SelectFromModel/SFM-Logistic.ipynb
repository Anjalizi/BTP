```python
import pandas as pd
from sklearn import model_selection, preprocessing
```

```python
data = pd.read_csv('parkinsons.csv', delimiter=',')
```

```python
x = data.drop(columns=['name', 'status'])
y = data['status']
```

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import LassoCV
```

```python
feature_names = x.columns
print(feature_names)
```

```python
import numpy as np
clf = LassoCV().fit(x, y)
importance = np.abs(clf.coef_)
print(importance)
```

```python
idx_third = importance.argsort()[-3]
threshold = importance[idx_third] + 0.001

idx_features = (-importance).argsort()[:2]
name_features = np.array(feature_names)[idx_features]
print('Selected features: {}'.format(name_features))

sfm = SelectFromModel(clf, threshold=threshold)
sfm.fit(x, y)
X_transform = sfm.transform(x)

n_features = sfm.transform(x).shape[1]
```

```python
# 70% training set and 30% testing set
x_train, x_test, y_train, y_test = model_selection.train_test_split (X_transform, y, test_size=0.3, random_state=0)
```

```python
x_val, x_test, y_val, y_test = model_selection.train_test_split (x_test, y_test, test_size=0.5, random_state=0)
```

```python
x_train = preprocessing.scale(x_train)
x_train.shape
```

```python
x_test = preprocessing.scale(x_test)
x_test.shape
```

```python
x_val = preprocessing.scale(x_val)
x_val.shape
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.linear_model import LogisticRegression
parameters = {'l1_ratio':(0,1), 'penalty':('l1', 'l2', 'elasticnet'), 'multi_class':('auto', 'ovr', 'multinomial'), 'C':range(1,10), 'solver':('newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga')}
# clf = LogisticRegression(random_state=0, max_iter=5000, C=1, l1_ratio=0, multi_class='multinomial',
#  penalty='l2', solver='newton-cg')
clf = LogisticRegression(random_state=0, max_iter=1000)
clf = GridSearchCV(clf, parameters)
```

```python
clf.fit(x_train, y_train)
```

```python
clf.best_params_
```

```python
clf.score(x_val, y_val)
```

```python
clf.score(x_test, y_test)
```
